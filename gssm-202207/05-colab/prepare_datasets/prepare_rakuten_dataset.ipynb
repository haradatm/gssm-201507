#### (2) 属性列の整形

```python
column_name = [
"カテゴリー",
"エリア",
"施設番号",
"施設名",
"コメント",
"総合",
"サービス",
"立地",
"部屋",
"設備・アメニティ",
"風呂",
"食事",
"旅行の目的",
"同伴者",
"宿泊年月",
"投稿者",
"年代",
"性別",
"投稿回数"
]
```

```python
def format_kuchikomi(path):
    datasets = []
    offset_attr = 5

    for line in open(path):
        attr_dic = {}
        cols = line.split('\t')
        # print(cols[5:])
        idx = offset_attr
        for i in range(0,20,2):
            attr_dic[cols[idx]] = cols[idx+1]
            idx += 2
        attr_dic['投稿者'] = cols[idx]
        attr_dic['年代'] = cols[idx+1]
        attr_dic['性別'] = cols[idx+2]
        attr_dic['投稿回数'] = cols[idx+3]

        # print('\t'.join(cols[0:offset_attr]), end='\t')
        record = cols[0:offset_attr]

        for i in range(offset_attr, offset_attr+10+4, 1):
            if column_name[i] in attr_dic:
                record.append(attr_dic[column_name[i]])
            else:
                record.append('na')

        # print(record)
        datasets.append(record)

    df_all = pd.DataFrame(datasets)
    df_all.columns = column_name
    return df_all
```

```python
df_all = format_kuchikomi("rakuten-all.tsv")
```
