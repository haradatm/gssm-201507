### 標準化残差行列Sを計算する

対応分析では変数間,個体間の距離を定義するために$\chi^2$距離を用いる

ここで $\chi^2$統計量は

$\displaystyle \chi^2 = \sum_{i,j}{\frac{(p_{ij}-r_{i} c_{i})^2}{r_{i} c_{i}}}$

となり, 相対度数 $p_{ij}$ の標準化残差の平方和である

### 手順

#### (1) 行と列の周辺度数を求める

$\displaystyle r_i = \Sigma_{j=1}^{j} p_{ij}  \quad  c_j = \Sigma_{i=1}^{i} p_{ij}$

#### (2) 行 r と列 c の対応行列を求める

$\displaystyle {D_r} = diag(r)  \quad  {D_c} = diag(c)$

$\displaystyle {D_r}^{-1/2} = diag(\frac{1}{\sqrt{r}})  \quad  {D_c}^{-1/2} = diag(\frac{1}{\sqrt{c}})$

#### (3) 標準化残差行列 S を計算する

$\displaystyle S = {D_r}^{-1/2} (P - rc^T) {D_c}^{-1/2}$

```r
r <- apply(table.P,1,sum) # 列の周辺度数の比率
c <- apply(table.P,2,sum) # 行の周辺度数の比率
Dr<-diag(r)  # r の対角行列
Dc<-diag(c)  # c の対角行列
Drmh<-diag(1/sqrt(r))  # Dr^(-1/2) を求める
Dcmh<-diag(1/sqrt(c))  # Dc^(-1/2) を求める
S <- Drmh %*% (data.matrix(table.P) - r %*% t(c)) %*% Dcmh  # S を求める

rownames(S)<-rownames(table.P)
colnames(S)<-colnames(table.P)
round(S[,1:10],digit=4)
```

### 試しに S を特異値分解で2次元にプロットする

$S=U D_{\alpha} V$

特異値分解の結果得られた $D_{\alpha}$ は固有値を成分とした対角行列.

n次元空間であった行列Pは,固有値ごとの軸に分解される.

```r
S.svd <- svd(S)

F <- S.svd$u
G <- S.svd$v

xlim <- c(min(F[,1], G[,1]), max(F[,1], G[,1])) * 1.1
ylim <- c(min(F[,2], G[,2]), max(F[,2], G[,2])) * 1.1

par(family = "HiraKakuProN-W3")
x<-F[,1]
y<-F[,2]
plot(x,y, col="red", pch=16, xlim=xlim, ylim=ylim)
text(x,y, col="red", rownames(table.P), adj=c(0,0))
par(new=T)
x<-G[,1]
y<-G[,2]
plot(x,y, col="grey", pch=16, xlim=xlim, ylim=ylim)
text(x,y, col="black", colnames(table.P), adj=c(1,1))
abline(h=0, v=0, lty="13")
```

### S を標準座標($\Phi$,$\Gamma$)で2次元にプロットする

対応分析では,固有ベクトルU,Vを$\frac{1}{\sqrt{r}}$や$\frac{1}{\sqrt{c}}$で基準化した値 $\Phi$, $\Gamma$ を用いる

$\displaystyle {D_r}^{-1/2} = diag(\frac{1}{\sqrt{r}})  \quad {D_c}^{-1/2} = diag(\frac{1}{\sqrt{c}})$

$\displaystyle \Phi = {D_r}^{-1/2} U$

$\displaystyle \Gamma = {D_c}^{-1/2} V$

```r
# 行,列ともに標準座標でプロットする

F <- Drmh %*% S.svd$u
G <- Dcmh %*% S.svd$v

xlim <- c(min(F[,1], G[,1]), max(F[,1], G[,1])) * 1.1
ylim <- c(min(F[,2], G[,2]), max(F[,2], G[,2])) * 1.1

par(family = "HiraKakuProN-W3")
x<-F[,1]
y<-F[,2]
plot(x,y, col="red", pch=16, xlim=xlim, ylim=ylim)
text(x,y, col="red", rownames(table.P), adj=c(0,0))
par(new=T)
x<-G[,1]
y<-G[,2]
plot(x,y, col="grey", pch=16, xlim=xlim, ylim=ylim)
text(x,y, col="black", colnames(table.P), adj=c(1,1))
abline(h=0, v=0, lty="13")
```

### S を主軸座標(F,G)で2次元にプロットする

$\displaystyle F = {D_r}^{-1/2} U D_{\alpha} = \Gamma D_{\alpha}$

$\displaystyle G = {D_c}^{-1/2} V D_{\alpha} = \Phi D_{\alpha}$

```r
# 行,列ともに主軸座標でプロットする

F <- Drmh %*% S.svd$u %*% diag(S.svd$d) 
G <- Dcmh %*% S.svd$v %*% diag(S.svd$d)

xlim <- c(min(F[,1], G[,1]), max(F[,1], G[,1])) * 1.1
ylim <- c(min(F[,2], G[,2]), max(F[,2], G[,2])) * 1.1

par(family = "HiraKakuProN-W3")
x<-F[,1]
y<-F[,2]
plot(x,y, col="red", pch=16, xlim=xlim, ylim=ylim)
text(x,y, col="red", rownames(table.P), adj=c(0,0))
par(new=T)
x<-G[,1]
y<-G[,2]
plot(x,y, col="grey", pch=16, xlim=xlim, ylim=ylim)
text(x,y, col="black", colnames(table.P), adj=c(1,1))
abline(h=0, v=0, lty="13")
```
