```python
from IPython.display import Image
%matplotlib inline
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
import numpy as np
import pandas as pd
```

```python
from urllib import request
from bs4 import BeautifulSoup

url = 'https://www.takahashi-w-clinic.jp/bbs.html'
response = request.urlopen(url)
soup = BeautifulSoup(response)
response.close()
```

```python
bbs = soup.find_all('article', class_='bbs')
print(len(bbs))
```

```python
qanda =  bbs[0].find_all('p')
print(len(qanda))
```

### カテゴリ

```python
category = [c.get_text() for c in qanda][0].replace('\t', '').split('\n')[2:]
category
```

```python
category = ",".join([s for s in [c.get_text() for c in qanda][0].replace('\t', '').split('\n')[2:] if s != ''])
category
```

### 相談内容

```python
question = [c.get_text() for c in qanda][1]
question
```

### 回答内容

```python
answer = [c.get_text() for c in qanda][2]
answer
```

```python
df = pd.DataFrame([question], columns=['question'])
df['answer'] = [answer]
df['category'] = [category]
df.head()
```

### 20件 (1ページ)

```python
q_list, a_list, c_list = [], [], []

for b in bbs:
    qanda =  b.find_all('p')
    c = ",".join([s for s in [c.get_text() for c in qanda][0].replace('\t', '').split('\n')[2:] if s != ''])
    q = [c.get_text() for c in qanda][1]
    a = [c.get_text() for c in qanda][2]
    q_list.append(q)
    a_list.append(a)
    c_list.append(c)
```

```python
df = pd.DataFrame(q_list, columns=['question'])
df['answer'] = a_list
df['category'] = c_list
df
```

### 1~470ページ分

```python
q_list, a_list, c_list = [], [], []

for i in range(470):
    i = i + 1
    if i == 1:
        url = 'https://www.takahashi-w-clinic.jp/bbs.html'
    else:
        url = f'https://www.takahashi-w-clinic.jp/bbs.html?page={i}'
    response = request.urlopen(url)
    soup = BeautifulSoup(response)
    response.close()
    bbs = soup.find_all('article', class_='bbs')

    for b in bbs:
        qanda =  b.find_all('p')
        texts = [c.get_text() for c in qanda]
        c = ",".join([s for s in texts[0].replace('\t', '').split('\n')[2:] if s != ''])
        q = texts[1] if len(texts) > 1 else ""
        a = texts[2] if len(texts) > 2 else ""
        q_list.append(q)
        a_list.append(a)
        c_list.append(c)
```

```python
df = pd.DataFrame(q_list, columns=['question'])
df['answer'] = a_list
df['category'] = c_list
df.head()
```

```python
df.to_csv('out.txt', sep='\t')
```
